## 테일러 급수

```python
import numpy as np
import matplotlib.pyplot as plt

# x 구간 설정
x = np.linspace(-2*np.pi, 2*np.pi, 500)

# 원래 함수
y_true = np.sin(x)

# 테일러 급수 근사 (1, 3, 5, 7차까지)
def taylor_sin(x, n_terms):
    result = 0
    for n in range(n_terms):
        term = ((-1)**n) * (x**(2*n + 1)) / np.math.factorial(2*n + 1)
        result += term
    return result

# 각각의 근사 함수 계산
y_1 = taylor_sin(x, 1)
y_3 = taylor_sin(x, 2)
y_5 = taylor_sin(x, 3)
y_7 = taylor_sin(x, 4)

# 그래프 시각화
labels = ['1st term', '3rd term', '5th term', '7th term']
results = [y_1, y_3, y_5, y_7]
fig, ax = plt.subplots(1, 4, figsize=(15, 6))
for i in range(4):
    ax[i].plot(x, y_true, label='sin(x)', color='black', linewidth=2)
    ax[i].plot(x, results[i], label=labels[i])
    ax[i].axhline(0, color='gray', linestyle='--', linewidth=0.5)
    ax[i].legend()
    ax[i].grid(True)
plt.title('sin(x) vs Taylor Series Approximations')
plt.show()
```
