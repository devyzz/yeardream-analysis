```python
'''
원형그래프 / plt.pie(수치, labels=라벨)
막대그래프 / plt.bar(x=라벨, heights=수치)
'''
labels = ["A", "B", "C", "D", "E"]
ratio = [33,25,17,17,8]

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
# 왼쪽: 파이차트
axes[0].pie(ratio, labels=labels)   #plt.pie(ratio,labels = labels)
axes[0].axis("equal")  # 원형으로 만들기
axes[0].set_title("원형 그래프")

# 오른쪽: 막대그래프
axes[1].bar(labels, ratio, color='skyblue') #plt.bar(labels, ratio)

axes[1].set_title("막대 그래프")

# 레이아웃 정리
plt.tight_layout()
plt.show()
```

```python
'''
히스토그램 - plt.hist() : 연속형 자료의 도수분포표를 기반으로 각 계급을 범주처럼 사용. 범주형 자료의 막대그래프와 같은 방식으로 그림
줄기-잎그림 - plt.stem(줄기, 관측 값) : 함수 사용시에 줄기 값을 따로 지정해줘야한다.
    ㄴ 장점 : 최대,최소값 등의 위치 파악이 용이함. 그리기 쉽다
    ㄴ 단점 : 관측값의 개수가 많은 경우 제한된공간내에 그릴 수 없다 / 관측값이 지나치게 흩어져잇으면 적절하지 못함
'''

drink_cup = pd.DataFrame({
    "cup": [22, 7, 19, 3, 10, 8, 19, 7, 15, 9, 35, 5], 
    "who": ["A", "E", "D", "B", "C", "A", "A", "A", "D", "B", "C", "B"],
    "stems": [2, 0, 1, 0, 1, 0, 1, 0, 1, 0, 3, 0]
})

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].hist(drink_cup["cup"]) #plt.hist(drink_cup["cup"]) # 구간별 빈도를 보여주는 그래프, "cup"값의 분포를 구간(bin)별로 나누어 시각화
axes[0].set_title("히스토그램")

axes[1].stem(drink_cup["stems"], drink_cup["cup"]) #stem(x,y)는 줄기 x축, 잎 y축으로 표현되는 그래프 
axes[1].set_title("줄기-잎 그래프")

plt.tight_layout()
plt.show()
```
