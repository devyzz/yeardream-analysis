```python
from math import comb

# 이항분포 확률 계산 함수
def binomial_probability(n, k, p):
    probability = comb(n, k) * (p ** k) * ((1 - p) ** (n - k))
    return probability

# 예제 조건 설정
n = 5  # 동전 던지기 횟수
k = 3  # 앞면이 나올 횟수
p = 0.5  # 앞면이 나올 확률

# 확률 계산
probability = binomial_probability(n, k, p)

# 결과 출력
print(f"동전을 {n}번 던졌을 때 앞면이 {k}번 나올 확률: {probability:.4f}")
```

```python
from math import comb

# 초기하분포 확률 계산 함수
def hypergeometric_probability(N, M, n, k):
    numerator = comb(M, k) * comb(N - M, n - k)
    denominator = comb(N, n)
    return numerator / denominator

# 예제 조건
N = 200  # 전체 제품 수
M = 20   # 불량품 수
n = 10   # 추출한 제품 수
k = 2    # 불량품이 2개 포함될 확률

# 확률 계산
probability = hypergeometric_probability(N, M, n, k)

# 결과 출력
print(f"총 {N}개 중 불량품 {M}개가 있는 상태에서 {n}개를 뽑아 불량품 {k}개를 뽑을 확률: {probability:.4f}")
```

```python
from math import exp, factorial

# 포아송분포 확률 계산 함수
def poisson_probability(lmbda, k):
    return (lmbda ** k) * exp(-lmbda) / factorial(k)

# 예제 조건
lmbda = 5  # 평균 발생률 (평균 전화 수)
k = 3      # 실제 발생 횟수 (전화 3번)

# 확률 계산
poisson_prob = poisson_probability(lmbda, k)

# 결과 출력
print(f"콜센터에 평균 {lmbda}건의 전화가 걸려올 때, 1시간 동안 전화가 {k}번 걸려올 확률: {poisson_prob:.4f}")
```

```python
from math import comb, exp, factorial

# 이항분포 확률 계산 함수
def binomial_probability(n, k, p):
    return comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

# 포아송분포 확률 계산 함수
def poisson_probability(lmbda, k):
    return (lmbda ** k) * exp(-lmbda) / factorial(k)

# 문제 조건
n = 10000    # 제품 수
p = 0.001    # 불량 확률
k = 10       # 불량품이 10개 나올 확률
lmbda = n * p  # 포아송 평균

# 이항분포 계산
binomial_prob = binomial_probability(n, k, p)

# 포아송분포 계산
poisson_prob = poisson_probability(lmbda, k)

# 결과 출력
print(f"이항분포로 계산한 확률: {binomial_prob:.4f}")
print(f"포아송분포로 계산한 확률: {poisson_prob:.4f}")
```

```python
from scipy.stats import uniform

# 연속형 균일분포 확률 계산 함수
def uniform_probability(a, b, x1, x2):
    # 확률 밀도 함수 계산
    density = 1 / (b - a)
    # 구간 확률 계산
    probability = (x2 - x1) * density
    return density, probability

# 문제 조건
a, b = 0, 10   # 전체 구간 [0, 10]
x1, x2 = 5, 7  # 관심 구간 [5, 7]

# 확률 계산
density, probability = uniform_probability(a, b, x1, x2)

# 결과 출력
print(f"균일분포의 확률 밀도 함수 값 (PDF): {density:.4f}")
print(f"구간 [{x1}, {x2}]에서 값이 나올 확률: {probability:.4f}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 정규분포 (평균 5, 표준편차 2)
x1 = np.linspace(-5, 15, 300)
y1 = norm.pdf(x1, 5, 2)

# 표준정규분포 (평균 0, 표준편차 1)
x2 = np.linspace(-5, 5, 300)
y2 = norm.pdf(x2, 0, 1)

# 그래프 그리기
plt.plot(x1, y1, label="Normal (μ=5, σ=2)", color="blue")
plt.plot(x2, y2, label="Standard Normal (μ=0, σ=1)", color="red", linestyle="--")
plt.title("Normal vs Standard Normal Distribution")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
from scipy.stats import norm

# Z-Score
z = 1.5

# 누적 분포 함수(CDF)를 이용한 확률 계산
probability = norm.cdf(z)
print(f"Z-Score가 1.5 이하일 확률: {probability:.4f}")
```

```python
import numpy as np
from scipy import stats

# 주어진 값
sample_mean = 105  # 표본평균
sigma = 10         # 모집단 표준편차
n = 15             # 표본 크기
confidence_level = 0.95  # 신뢰수준

# 신뢰계수 Z 값
alpha = 1 - confidence_level
z_value = stats.norm.ppf(1 - alpha / 2)

# 표준오차 계산
standard_error = sigma / np.sqrt(n)

# 오차 한계 계산
margin_of_error = z_value * standard_error

# 신뢰구간 계산
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"표본평균: {sample_mean}")
print(f"95% 신뢰구간: ({lower_bound:.2f}, {upper_bound:.2f})")
```
