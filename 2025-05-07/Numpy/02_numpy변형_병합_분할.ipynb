## 1ï¸âƒ£ ë°°ì—´ì˜ í˜•íƒœ ë°”ê¾¸ê¸° - `reshape()`

```python
import numpy as np

a = np.arange(12)
print("ì›ë³¸ ë°°ì—´:", a)

reshaped = a.reshape(3, 4)
print("3x4ë¡œ reshape:\n", reshaped)

# reshapeì˜ ì£¼ìš” ì¡°ê±´:
# - ì „ì²´ ìš”ì†Œ ìˆ˜ëŠ” ë™ì¼í•´ì•¼ í•¨
# - (12,) â†’ (3,4), (4,3), (2,6) ë“± ê°€ëŠ¥
```

## 2ï¸âƒ£ ë‹¤ì°¨ì› ë°°ì—´ì„ 1ì°¨ì›ìœ¼ë¡œ - `flatten()` / `ravel()`

```python
b = np.array([[1, 2, 3], [4, 5, 6]])

flat1 = b.flatten()  # ë³µì‚¬ë³¸ ë°˜í™˜
flat2 = b.ravel()    # ì›ë³¸ view ë°˜í™˜

print("flatten:", flat1)
print("ravel:", flat2)
```

## 3ï¸âƒ£ ë°°ì—´ ì „ì¹˜ - `transpose()` / `.T`

```python
c = np.array([[1, 2, 3], [4, 5, 6]])

print("ì›ë³¸:", c)
print("ì „ì¹˜ (transpose):", c.transpose())
print("ì „ì¹˜ (.T ì†ì„±):", c.T)
```

## 4ï¸âƒ£ ë°°ì—´ ì´ì–´ë¶™ì´ê¸° - `concatenate()`

```python
a1 = np.array([[1, 2], [3, 4]])
a2 = np.array([[5, 6]])

# í–‰ ê¸°ì¤€ ì´ì–´ë¶™ì´ê¸° (axis=0)
combined = np.concatenate((a1, a2), axis=0)
print("í–‰ ê¸°ì¤€ ì´ì–´ë¶™ì´ê¸°:", combined)

# ì—´ ê¸°ì¤€ ì´ì–´ë¶™ì´ê¸° (axis=1)
a3 = np.array([[7], [8]])
combined2 = np.concatenate((a1, a3), axis=1)
print("ì—´ ê¸°ì¤€ ì´ì–´ë¶™ì´ê¸°:", combined2)
```

## 5ï¸âƒ£ ë°°ì—´ ìŒ“ê¸° - `stack()`, `vstack()`, `hstack()`

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print("stack:", np.stack((x, y)))         # axis=0ì´ ê¸°ë³¸
print("vstack:", np.vstack((x, y)))       # ìˆ˜ì§ìœ¼ë¡œ ìŒ“ê¸°
print("hstack:", np.hstack((x, y)))       # ìˆ˜í‰ìœ¼ë¡œ ìŒ“ê¸°
```
