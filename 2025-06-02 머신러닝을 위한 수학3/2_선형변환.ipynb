## 직교행렬

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm

font_path = 'C:\\Windows\\Fonts\\NGULIM.TTF'
font = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font)

# 회전각 (30도)
theta = np.deg2rad(30)

# 2D 직교 회전 행렬
Q = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])

# 단위벡터들
x = np.array([1, 0])
y = np.array([0, 1])

# 회전된 벡터들
x_rot = Q @ x
y_rot = Q @ y

# 시각화
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original X')
ax[0].quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1, color='green', label='Original Y')
ax[0].set_title("기존 벡터")

ax[1].quiver(0, 0, x_rot[0], x_rot[1], angles='xy', scale_units='xy', scale=1, color='cyan', label='Rotated X')
ax[1].quiver(0, 0, y_rot[0], y_rot[1], angles='xy', scale_units='xy', scale=1, color='lime', label='Rotated Y')
ax[1].set_title("직교행렬에 의해 변환된 벡터")

for i in range(2):
    ax[i].set_xlim(-1.5, 1.5)
    ax[i].set_ylim(-1.5, 1.5)
    ax[i].grid(True)
    ax[i].legend()
    
plt.show()
```
