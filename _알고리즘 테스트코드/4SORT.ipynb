```python
def insertion_sort(arr):
    print("초기 배열:", arr)
    print("정렬 과정:")
    for i in range(1, len(arr)):
        key = arr[i]  # 현재 삽입할 값
        j = i - 1

        print(f"\n[{i}] key = {key} → 앞쪽에 삽입 위치 찾기")

        # key보다 큰 값은 오른쪽으로 밀기
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            print(f"  arr[{j}] = {arr[j]} > key → 오른쪽으로 이동: {arr}")
            j -= 1

        arr[j + 1] = key  # key를 올바른 위치에 삽입
        print(f"  key 삽입 → arr[{j + 1}] = {key}: {arr}")

    return arr

# 실행
arr = [5, 3, 1, 4, 2]
sorted_arr = insertion_sort(arr.copy())
print("\n삽입 정렬 결과:", sorted_arr)

```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        print(f"\n{i+1}회전 시작: {arr}")  # 각 회전 시작 전 출력

        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                # 스왑 (오름차순 정렬)
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                print(f"  → {arr[j]}와 {arr[j+1]} 교환 → {arr}")
            else:
                print(f"  → {arr[j]}와 {arr[j+1]} 그대로")

    return arr

# 실행
data = [5, 3, 1, 4, 2]
print("정렬 전:", data)
sorted_data = bubble_sort(data.copy())
print("\n정렬 후:", sorted_data)
```

```python
def merge_sort(arr, depth=0):
    indent = "  " * depth  # 재귀 깊이만큼 들여쓰기
    print(f"{indent}정렬할 배열: {arr}")

    if len(arr) <= 1:
        print(f"{indent}반환: {arr}")
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid], depth + 1)
    right = merge_sort(arr[mid:], depth + 1)

    merged = merge(left, right, depth)
    print(f"{indent}병합 결과: {merged}")
    return merged

def merge(left, right, depth):
    indent = "  " * depth
    print(f"{indent}병합 중: {left} + {right}")
    result = []
    i = j = 0

    # 작은 값부터 차례로 추가
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # 남은 값들 추가
    result.extend(left[i:])
    result.extend(right[j:])
    return result

# 테스트 실행
arr = [5, 3, 1, 4, 2]
print("초기 배열:", arr)
sorted_arr = merge_sort(arr)
print("\n합병 정렬 결과:", sorted_arr)
```

```python
# merge_sort()-->배열을 계속 반으로 나누는 함수-->재귀 호출, 분할(divide)
# merge()-->나눈 배열을 정렬하며 다시 합치는 함수-->병합(merge), 정렬하며 결합(combine)
def merge_sort(arr, depth=0):
    print(f"[정렬 시작] 배열: {arr}")  # 각 정렬 시작 시 출력

    if len(arr) <= 1:
        print(f"→ 반환 (정렬 완료): {arr}\n")  # 반환 시 구분 출력
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid], depth + 1)     # 왼쪽 재귀
    right = merge_sort(arr[mid:], depth + 1)    # 오른쪽 재귀

    print(f"병합 시작: {left} + {right}")     # 병합 시작 표시
    merged = merge(left, right)
    print(f"병합 완료: {merged}")             # 병합 완료 표시
    print("—" * 40)                              # 구분선으로 단계 구분
    return merged

def merge(left, right):
    result = []
    i = j = 0

    # 병합 처리: 두 리스트에서 작은 값부터 꺼냄
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # 남은 요소 추가
    result.extend(left[i:]), result.extend(right[j:])
    return result

# 실행
arr = [5, 3, 1, 4, 2]
print("초기 배열:", arr),print("=" * 40)
sorted_arr = merge_sort(arr)
print("=" * 40),print("최종 정렬 결과:", sorted_arr)
```

```python
def quick_sort(arr, depth=0):
    indent = "  " * depth  # 재귀 깊이에 따라 들여쓰기
    print(f"{indent}정렬할 배열: {arr}")

    if len(arr) <= 1:
        print(f"{indent}반환: {arr}")
        return arr

    pivot = arr[0]  # 첫 번째 값을 피벗으로 선택
    left = [x for x in arr[1:] if x < pivot]
    right = [x for x in arr[1:] if x >= pivot]

    print(f"{indent}피벗: {pivot}")
    print(f"{indent}작은 값들(left): {left}")
    print(f"{indent}큰 값들(right): {right}")

    sorted_left = quick_sort(left, depth + 1)
    sorted_right = quick_sort(right, depth + 1)

    result = sorted_left + [pivot] + sorted_right
    print(f"{indent}병합 결과: {result}")
    return result

# 실행 예제
arr = [5, 3, 1, 4, 2]
print("초기 배열:", arr)
sorted_arr = quick_sort(arr)
print("\n퀵 정렬 결과:", sorted_arr)
```

```python
compare_count = 0             # 총 비교 횟수를 저장하는 변수
pivot_history = []            # 피벗 값들이 선택된 순서를 저장하는 리스트

def quick_sort(arr):
    global compare_count, pivot_history  # 전역 변수 사용 선언
    print(f"정렬할 배열: {arr}\n")        

    if len(arr) <= 1:                     # 배열 길이가 0 또는 1이면 이미 정렬된 상태
        print(f"반환: {arr}")             # 그대로 반환
        return arr

    pivot = arr[0]                        # 피벗: 배열의 첫 번째 요소 선택
    pivot_history.append(pivot)          # 선택된 피벗을 기록 리스트에 저장

    left = []                             # 피벗보다 작은 값을 담을 리스트
    right = []                            # 피벗보다 크거나 같은 값을 담을 리스트

    for x in arr[1:]:                     # 피벗을 제외한 나머지 값들 반복
        compare_count += 1               # 비교 횟수 1 증가
        if x < pivot:                     # 피벗보다 작으면
            left.append(x)               # left 쪽에 추가
        else:
            right.append(x)              # 크거나 같으면 right 쪽에 추가

    print(f"피벗: {pivot}")               # 현재 선택된 피벗 출력
    print(f"작은 값들: {left}") , print(f"큰 값들: {right}")           
    sorted_left = quick_sort(left)
    sorted_right = quick_sort(right) #왼쪽/오른쪽 부분 재귀 정렬

    result = sorted_left + [pivot] + sorted_right  # 병합: 왼쪽 + 피벗 + 오른쪽
    print(f"병합 결과: {result}")         
    return result                        

# 실행 부분
arr = [5, 3, 1, 4, 2]
print("초기 배열:", arr)  # 시작 메시지 출력

sorted_arr = quick_sort(arr)             # 퀵 정렬 실행 및 결과 저장
print("\n최종 정렬 결과:", sorted_arr)  # 최종 정렬 결과 
print(f"총 비교 횟수: {compare_count}")  # 총 비교 횟수 
print(f"피벗 선택 순서: {pivot_history}") # 피벗 선택 순서 

```

```python
def heapify(arr, n, i, depth=0):
    indent = "  " * depth
    largest = i      # 현재 노드를 최대값이라고 가정
    left = 2 * i + 1 # 왼쪽 자식
    right = 2 * i + 2 # 오른쪽 자식
    print(f"{indent}heapify 호출: index={i}, 값={arr[i]}")

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        print(f"{indent}→ 교환: arr[{i}]={arr[i]} <-> arr[{largest}]={arr[largest]}")
        arr[i], arr[largest] = arr[largest], arr[i]  # 스왑
        heapify(arr, n, largest, depth + 1)  # 재귀적으로 자식 노드 heapify

def heap_sort(arr):
    n = len(arr)
    print("원본 배열:", arr)
    print("\n[1단계] 최대 힙 구성 ----------------------")
    
    # 힙 구성 (완전 이진트리의 마지막 부모부터)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
        print(f"→ 현재 힙: {arr}")

    print("\n[2단계] 정렬 과정 --------------------------")

    # 정렬
    for i in range(n - 1, 0, -1):
        print(f"\n→ 최대값 {arr[0]}을 끝으로 이동 (arr[{i}])")
        arr[0], arr[i] = arr[i], arr[0]  # 가장 큰 값 맨 뒤로
        print(f"  교환 후: {arr}")
        heapify(arr, i, 0)  # 남은 힙을 재정렬
        print(f"  힙 재정렬 후: {arr}")

    return arr

arr = [5, 3, 1, 4, 2]
sorted_arr = heap_sort(arr.copy())
print("\n정렬 결과:", sorted_arr)
```

```python
def heapify(arr, n, i):
    # 현재 노드를 루트로 하는 서브트리에서 최대 힙을 유지
    largest = i            # 루트 노드
    left = 2 * i + 1       # 왼쪽 자식
    right = 2 * i + 2      # 오른쪽 자식

    # 왼쪽 자식이 루트보다 크면 largest를 왼쪽 자식으로
    if left < n and arr[left] > arr[largest]:
        largest = left

    # 오른쪽 자식이 largest보다 크면 largest를 오른쪽 자식으로
    if right < n and arr[right] > arr[largest]:
        largest = right

    # largest가 루트가 아니면 스왑 후, 재귀적으로 heapify 수행
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)
    print("▶ 원본 배열:", arr)

    # 1단계: 최대 힙 구성
    print("\n[1단계] 최대 힙 구성")
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
        print(f"  heapify(i={i}) → {arr}")

    # 2단계: 힙 정렬
    print("\n[2단계] 힙 정렬 시작")
    for i in range(n - 1, 0, -1):
        # 현재 최대값 (루트)을 맨 뒤로 보내기
        arr[0], arr[i] = arr[i], arr[0]
        print(f"  최대값 {arr[i]}을 끝으로 이동: {arr}")
        # 남은 배열에서 힙 속성 유지
        heapify(arr, i, 0)
        print(f"  heapify 후: {arr}")

    return arr

# 실행 예제
arr = [5, 3, 1, 4, 2]
sorted_arr = heap_sort(arr.copy())
print("\n✅ 정렬 결과:", sorted_arr)
```

```python
def heapify(arr, n, i):
    # i번 인덱스를 루트로 하는 트리를 힙 조건에 맞게 정리
    largest = i          # 현재 루트를 가장 큰 값이라 가정
    left = 2 * i + 1     # 왼쪽 자식 인덱스
    right = 2 * i + 2    # 오른쪽 자식 인덱스

    # 왼쪽 자식이 더 크면 largest 변경
    if left < n and arr[left] > arr[largest]:
        largest = left

    # 오른쪽 자식이 더 크면 largest 변경
    if right < n and arr[right] > arr[largest]:
        largest = right

    # 가장 큰 값이 현재 노드가 아니면 자식과 교환하고 다시 heapify
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)
    print("정렬 전 배열:", arr)

    # 1단계: 최대 힙 만들기
    print("\n[1단계] 최대 힙 만들기")
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
        print(f"  {i}번 인덱스 기준으로 힙 구성: {arr}")

    # 2단계: 정렬 시작
    print("\n[2단계] 힙 정렬 시작")
    for i in range(n - 1, 0, -1):
        # 루트(최댓값)를 끝으로 보냄
        arr[0], arr[i] = arr[i], arr[0]
        print(f"  최댓값 {arr[i]}을 끝으로 이동 → {arr}")
        # 다시 힙 정리 (맨 끝은 제외)
        heapify(arr, i, 0)
        print(f"  힙 재정리 결과 → {arr}")

    print("\n정렬 완료!")
    return arr

# 실행
arr = [5, 3, 1, 4, 2]
sorted_arr = heap_sort(arr.copy())
print("\n✅ 최종 정렬 결과:", sorted_arr)
```

```python
def heapify(arr, n, i):                                # 힙 성질을 유지하도록 i를 루트로 하는 트리를 정리
    largest = i                                         # 현재 루트를 가장 큰 값이라 가정
    left = 2 * i + 1                                    # 왼쪽 자식 인덱스
    right = 2 * i + 2                                   # 오른쪽 자식 인덱스

    if left < n and arr[left] > arr[largest]:           # 왼쪽 자식이 루트보다 크면 largest 변경
        largest = left

    if right < n and arr[right] > arr[largest]:         # 오른쪽 자식이 largest보다 크면 largest 변경
        largest = right

    if largest != i:                                    # largest가 루트가 아니면
        arr[i], arr[largest] = arr[largest], arr[i]     # 루트와 가장 큰 자식 값 교환
        heapify(arr, n, largest)                        # 바뀐 자식을 기준으로 다시 heapify (재귀 호출)

def heap_sort(arr):                                     # 힙 정렬 함수 시작
    n = len(arr)                                        # 배열 길이 저장
    print("정렬 전 배열:", arr)                         # 원본 배열 출력

    print("\n[1단계] 최대 힙 만들기")                   # 최대 힙 만들기 시작
    for i in range(n // 2 - 1, -1, -1):                  # 마지막 부모부터 시작해서 루트까지 heapify
        heapify(arr, n, i)                              # i를 루트로 하는 트리를 정리
        print(f"  {i}번 인덱스 기준으로 힙 구성: {arr}")  # 중간 힙 상태 출력

    print("\n[2단계] 힙 정렬 시작")                      # 정렬 시작
    for i in range(n - 1, 0, -1):                        # 배열 끝에서부터 하나씩 줄여가며 정렬
        arr[0], arr[i] = arr[i], arr[0]                 # 루트(최댓값)과 마지막 원소 교환
        print(f"  최댓값 {arr[i]}을 끝으로 이동 → {arr}") # 최대값이 뒤로 이동한 상태 출력
        heapify(arr, i, 0)                              # 남은 배열에 대해 다시 힙 정리
        print(f"  힙 재정리 결과 → {arr}")               # 재정리된 힙 출력

    print("\n정렬 완료!")                                # 정렬 완료 메시지 출력
    return arr                                           # 정렬된 배열 반환

# 실행 부분
arr = [5, 3, 1, 4, 2]                                    # 정렬할 원본 배열
sorted_arr = heap_sort(arr.copy())                      # 복사본을 힙 정렬하여 결과 저장
print("\n 최종 정렬 결과:", sorted_arr)               # 최종 결과 출력

```
